```python
import pandas as pd
```

```python
# Reading COVID world data into a dataframe
covid_df = pd.read_csv("covid_data.csv")
covid_df
```

```python
#Selecting only the necessary columns for our model
covid_filtered_df = covid_df[['location','date','people_vaccinated','people_fully_vaccinated','population','population_density','gdp_per_capita']]
covid_filtered_df
```

```python
covid_filtered_df.dropna(inplace = True)
```

```python
covid_count = covid_filtered_df['location'].value_counts()
covid_count
```

```python
covid_filtered2= covid_filtered_df.location.isin(covid_count[covid_count>1].index)
covid_filtered2
```

```python
#Gathering countries with multiple COVID entries
covid_filtered_df= covid_filtered_df[covid_filtered2]
covid_filtered_df
```

```python
covid_filtered_df['location'].unique()
```

```python
covid_filtered_df= covid_filtered_df[covid_filtered_df.location != 'World']
covid_filtered_df
```

```python
#Creating new column for vaccination percentage 
covid_filtered_df["vaccination_percentage"] = covid_filtered_df["people_fully_vaccinated"] / covid_filtered_df["population"] * 100
covid_filtered_df
```

```python
#Gathering 3 countries' data for test model
model_countries= ['Israel', 'United Kingdom', 'United States']

model_countries_list= covid_filtered_df.location.isin(model_countries)

covid_test_df= covid_filtered_df[model_countries_list]
covid_test_df
```

```python
covid_test_df.to_csv('covid_test_model_data.csv', index=False)
```
